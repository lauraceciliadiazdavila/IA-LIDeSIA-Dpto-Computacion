## Funciones


*   def
*   lambda



```python
# Suma dos numeros y devuelve un numero con la suma de los argumentos
def suma(a, b):
    return a + b

print(suma(2, 3))
```

Python es indentado significa que los bloques de código se definen por la sangría (espacios al inicio de la línea)

Ver los espacios en "return" luego de los ":" de la función

```python
# Con anotaciones de tipo
def area_rect(w: float, h: float) -> float:
    return w * h

print(area_rect(3.0, 2.0))
```

```python
# Lambda simple
doble = lambda x: x * 2
print(doble(7))
```

```python
# Ternario en lambda (una sola expresión)
clasificar = lambda x: "es par" if x % 2 == 0 else "impar"
print(clasificar(7))
```

## Ciclos de repetición


```python
# Recorre una lista
numeros = [1, 2, 3, 4, 5]
suma = 0
for n in numeros:
    suma += n
print("suma:", suma)
```

```python
# Repite mientras la condición sea verdadera
contador = 3
while contador > 0:
    print("tick", contador)
    contador -= 1
```

## Numpy

Importamos el módulo de computación numérica más usado en python: [numpy](https://numpy.org/)

Para poder utilizarlo lo hacemos con la instrucción "import"

```python
import numpy as np

# Crear un arreglo Numpy
arreglo = np.array([1, 2, 3, 4, 5])

# Calcular la media del arreglo
media = np.mean(arreglo)

print("Arreglo:", arreglo)
print("Media:", media)
```

[Matplotlib](https://matplotlib.org) es una librería de Python para crear gráficos y visualizaciones de datos.

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0, 10, 100)        # crea un arreglo de 100 puntos entre 0 y 10
y = np.sin(x)                      # calcula el seno de cada punto

# Crear gráfico
plt.plot(x, y, label="Seno", color="blue")

# Personalizar
plt.title("Gráfico de Seno usando NumPy y Matplotlib")
plt.xlabel("Eje X")
plt.ylabel("Eje Y")
plt.legend()
plt.grid(True)

# Mostrar
plt.show()
```

## Pandas

[Pandas](https://pandas.pydata.org/) es una librería muy utilizada en python para manipular datos. El formato más utilizado es el de Dataframe y se podría decir que es el Excel de Python, podemos manipular datos en formato de tabla.

```python
import pandas as pd

# Datos de ejemplo
data = {
    "alumno":  ["Ana", "Luis", "Mara", "Juan", "Sofía"],
    "grupo":   pd.Series([1, 1, 2, 2, 3]),
    "nota_p1": [8, 9, 7, 6, 10],
    "nota_p2": [9, 8, 8, 7, 9],
}

# Se genera df a partir de un diccionario
df = pd.DataFrame(data)

# Columna nueva: promedio por alumno
df["promedio"] = df[["nota_p1", "nota_p2"]].mean(axis=1)

# Ver primeras filas
print(df.head())

#Filtrar: alumnos con promedio >= 8
print(df[df["promedio"] >= 8][["alumno", "promedio"]])

#Agrupar: promedio por grupo
print(df.groupby("grupo")["promedio"].mean().reset_index())

#Ordenar: top 3 promedios
print(df.sort_values("promedio", ascending=False).head(3))
```

```python
df.describe()
```

pd.Series es el objeto básico de pandas para representar una secuencia de datos unidimensional (parecido a un array de NumPy o una columna de Excel), con un índice que etiqueta cada elemento.

Es una sola columna de datos, con un valor y su etiqueta (índice).

```python
s = pd.Series([10, 20, 30], index=["a", "b", "c"])
print(s)
```

## PIL

PIL/Pillow es la librería estándar en Python para trabajar con imágenes de forma sencilla y soporta muchos formatos y operaciones básicas de edición.

```python
from PIL import Image
import numpy as np

# Crear una imagen de ejemplo y guardarla como PNG
image_array = np.random.randint(0, 256, (100, 100, 3), dtype=np.uint8) # genero arreglo np (100, 100, 3) RGB
image = Image.fromarray(image_array, 'RGB')
png_path = 'example_image.png'
image.save(png_path)
png_path
```
